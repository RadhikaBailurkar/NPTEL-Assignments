```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
from scipy import stats
```

```python
df = pd.read_excel("demo.xlsx")
df
```

```python
fvalue, pvalue = stats.f_oneway(df['Schedule1'], df['Schedule2'], df['Schedule3'])
print(fvalue, pvalue)
```

# ANOVA

```python
data = pd.melt(df.reset_index(),id_vars = ['index'], value_vars=['Schedule1','Schedule2','Schedule3'])
data.columns = ['index','treatments','value']
     
```

```python
data
```

```python
model = ols('value ~ C(treatments)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=1)
anova_table
```

```python
#Accept the NULL hypo, bcoz P>0.05
```
